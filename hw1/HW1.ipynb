## **Section 2: Measure of Power (Q2a & 2b)**

## 2a
**Using this fields:**
- Census_ProcessorCoreCount: Number of processor cores.
- Census_PrimaryDiskTotalCapacity: Total disk capacity (MB).
- Census_TotalPhysicalRAM: Total physical RAM (MB).

```python
power_measure_col = [
    'Census_ProcessorCoreCount',
    'Census_PrimaryDiskTotalCapacity',
    'Census_TotalPhysicalRAM',
    'HasDetections'
]

measure_power = df[power_measure_col].copy()
measure_power.head()
```

```python
measure_power['Census_PrimaryDiskTotalCapacity'].replace(0, 1, inplace=True)

weight_core = 0.4
weight_disk = 0.3
weight_RAM = 0.3

# Normalize values

measure_power['Normalized_Core_Count'] = measure_power['Census_ProcessorCoreCount'] / measure_power['Census_ProcessorCoreCount'].max()
measure_power['Log_PrimaryDiskTotalCapacity'] = np.log1p(measure_power['Census_PrimaryDiskTotalCapacity'])
measure_power['Normalized_Disk'] = measure_power['Log_PrimaryDiskTotalCapacity'] / measure_power['Log_PrimaryDiskTotalCapacity'].max()
#measure_power['Normalized_Disk'] = measure_power['Census_PrimaryDiskTotalCapacity'] / measure_power['Census_PrimaryDiskTotalCapacity'].max()
measure_power['Normalized_RAM'] = measure_power['Census_TotalPhysicalRAM'] / measure_power['Census_TotalPhysicalRAM'].max()

# Calculate Power Index

measure_power['Power_Index'] = (weight_core * measure_power['Normalized_Core_Count'] + weight_disk * measure_power['Normalized_Disk'] + weight_RAM * measure_power['Normalized_RAM'])

measure_power.head(15)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(measure_power['Power_Index'], bins=10)
plt.title('Distribution of Computer Power Index')
plt.xlabel('Power Index')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='HasDetections', y='Power_Index', data=measure_power)
plt.title('Power Index vs Malware Detection (Is Protected)')
plt.xlabel('Has Detections (1: Yes, 0: No)')
plt.ylabel('Power Index')
plt.xticks([0, 1], ['Has Malware', 'No Malware'])
plt.show()
```

## Answer
By observing the plot we can see that the data is similarly distributed, we can see that some of the more powerful computers contain some malware, we can conclude that the power of the computer cannot tell if a computer is more secure or not.

## **Section 3: OS version vs Malware detected (Q3)**

```python
# Count malware detections based on Census_OSBuildNumber
build_counts = df.groupby('Census_OSBuildNumber')['HasDetections'].value_counts().unstack(fill_value=0)

# Calculate the percentage of detections
build_counts['Total'] = build_counts.sum(axis=1)
build_counts['Malware_Detection_Percentage'] = (build_counts[1] / build_counts['Total']) * 100

# Count malware detections based on Census_OSBuildRevision
revision_counts = df.groupby('Census_OSBuildRevision')['HasDetections'].value_counts().unstack(fill_value=0)

# Calculate the percentage of detections
revision_counts['Total'] = revision_counts.sum(axis=1)
revision_counts['Malware_Detection_Percentage'] = (revision_counts[1] / revision_counts['Total']) * 100

build_counts.head()
```

```python
plt.figure(figsize=(15, 5))

# Plot number of detections
plt.subplot(1, 2, 1)
build_counts[1].plot(kind='bar', color='red', alpha=0.6, label='Malware Detections', stacked=True)
build_counts[0].plot(kind='bar', color='blue', alpha=0.6, label='No Malware', stacked=True)
plt.title('Malware Detections by OS Build Number')
plt.xlabel('OS Build Number')
plt.ylabel('Count')
plt.xticks(rotation=45)
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::5])
plt.legend()
plt.grid()

# Plot percentage of detections
plt.subplot(1, 2, 2)
build_counts['Malware_Detection_Percentage'].plot(kind='bar', color='purple', alpha=0.6)
plt.title('Malware Detection Percentage by OS Build Number')
plt.xlabel('OS Build Number')
plt.ylabel('Detection Percentage (%)')
plt.xticks(rotation=45)
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::5])
plt.grid()
plt.tight_layout()
plt.show()

# Plot for Census_OSBuildRevision
plt.figure(figsize=(15, 5))

# Plot number of detections
plt.subplot(1, 2, 1)
revision_counts[1].plot(kind='bar', color='red', alpha=0.6, label='Malware Detections')
revision_counts[0].plot(kind='bar', color='blue', alpha=0.6, label='No Malware')
plt.title('Malware Detections by OS Build Revision')
plt.xlabel('OS Build Revision')
plt.ylabel('Count')
plt.xticks(rotation=45)
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::8])
plt.legend()
plt.grid()

# Plot percentage of detections
plt.subplot(1, 2, 2)
revision_counts['Malware_Detection_Percentage'].plot(kind='bar', color='purple', alpha=0.6)
plt.title('Malware Detection Percentage by OS Build Revision')
plt.xlabel('OS Build Revision')
plt.ylabel('Detection Percentage (%)')
plt.xticks(rotation=45)
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::8])
plt.grid()

plt.tight_layout()
plt.show()
```

## Answer

When examining the data from OS build number, the smaller build number tend to have a bigger percentage of detected malware.

While looking at the data for OS build revision, we are able to see that higher revision numbers have a bigger percentage in detection.
